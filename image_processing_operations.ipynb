### Reading and Showing image using OpenCV

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('./Images/image1.jpeg')
img2 = cv2.imread('./Images/image2.jpeg')
img = cv2.imread('./Images/all_done.jpeg', 1)

print(img.shape)
h,w,c = img.shape
print("Dimensions of the image is:nnHeight:", h, "pixelsnWidth:", w, "pixelsnNumber of Channels:", c)
print(type(img))
print(img.dtype)
```

```python
plt.imshow(img1)
plt.show()
plt.imshow(img2)
```

### merge images and substract images

```python
merge_image = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
sub = cv2.subtract(img1, img2)

plt.imshow(merge_image)
plt.show()
plt.imshow(sub)
```

### Resizing image

```python

 
half = cv2.resize(img1, (0, 0), fx = 0.5, fy = 0.5)
cv2.imwrite('half.jpeg', half)
bigger = cv2.resize(img1, (1050, 1610)) 
stretch_near = cv2.resize(img1, (780, 540), interpolation = cv2.INTER_NEAREST) 
 
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[img1, half, bigger, stretch_near]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
plt.show()
```

### find edges from image

```python
import cv2
  
# Setting parameter values
t_lower = 50  # Lower Threshold
t_upper = 150  # Upper threshold

edge = cv2.Canny(img1, t_lower, t_upper, apertureSize = 3, L2gradient = True)
  
plt.imshow(img1)
plt.show()
plt.imshow(edge)
```

```python
img = cv2.imread("./Images/all_done.jpeg")  # Read image
Gaussian = cv2.GaussianBlur(img, (25,25), 0)
plt.imshow(Gaussian)
```

```python
kernel = np.ones((5,5), np.uint8)
img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)
plt.imshow(img)
plt.show()
plt.imshow(img_erosion)
plt.show()
plt.imshow(img_dilation)
```

### Image Threshold

```python
img = cv2.imread("./Images/all_done.jpeg",0)
#gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("original",img)
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)

cv2.imshow("binary Threshold",thresh1)
cv2.imshow("binary Threshold inverted",thresh2)
cv2.imshow("Truncated Threshold",thresh3)
cv2.imshow("set to 0",thresh4)
cv2.imshow("set to 0 inverted",thresh5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Image Threshold mean gaussian OTSU

```python
img = cv2.imread("./Images/book.jpeg",0)
thresh1 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,199,5)
thresh2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,199,5)
ret, thresh = cv2.threshold(img,120,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
cv2.imshow("adaptive thresh mean",thresh1)
cv2.imshow("adaptive thresh gaussian",thresh2)
cv2.imshow("OTSU THRESHOLD",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### detect moving object

```python
cap = cv2.VideoCapture('./video/camera.mp4')
fgbg = cv2.createBackgroundSubtractorMOG2()
 
while(1):
    ret, frame = cap.read()
    if ret == False:
       break
    fgmask = fgbg.apply(frame)

    cv2.imshow('fgmask', fgmask)
    cv2.imshow('frame',frame )

    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
 
cap.release()
cv2.destroyAllWindows()
```

### Create frames from video

```python
# Importing all necessary libraries
import cv2
import os
  
# Read the video from specified path
cam = cv2.VideoCapture("./video/camera.mp4")
  
try:
      
    # creating a folder named data
    if not os.path.exists('data'):
        os.makedirs('data')
  
# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')
  
# frame
currentframe = 0
  
while(True):
      
    # reading from frame
    ret,frame = cam.read()
  
    if ret:
        # if video is still left continue creating images
        name = './data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)
  
        # writing the extracted images
        cv2.imwrite(name, frame)
  
        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break
  
# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
